```python
import os
import pandas as pd
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import numpy as np

# Correct your folder name
image_folder = 'image_dataset'

# Define target image size
image_size = (128, 128)

# Load your labels CSV
labels_df = pd.read_csv('image_labels.csv')  # Or whatever your CSV is called
print(labels_df.head())
# Initialize lists
images = []
labels = []

# Load images and their corresponding labels based on the CSV
for idx, row in labels_df.iterrows():
    img_name = row['filename']  # Filename (e.g., img_1.png)
    label = row['label']  # Label (either 0 or 1)
    img_path = os.path.join(image_folder, img_name)


    # Check if the image exists
    if os.path.exists(img_path):
        img = load_img(img_path, target_size=image_size)
        img_array = img_to_array(img) / 255.0
        images.append(img_array)
        labels.append(label)  # Append label for corresponding image
    else:
       print(f"Image {img_name} not found at {img_path}")

# Convert to numpy arrays
images = np.array(images)
labels = np.array(labels)

# Print shapes to confirm
print(f"Images shape: {images.shape}, Labels shape: {labels.shape}")
```

```python
# Split into 70% train and 30% test
X_train_70, X_test_30, y_train_70, y_test_30 = train_test_split(
    images, labels, test_size=0.30, random_state=42, stratify=labels
)

print(f"70/30 Split - Train shape: {X_train_70.shape}, Test shape: {X_test_30.shape}")
```

```python
# Split into 50% train and 50% test
X_train_50, X_test_50, y_train_50, y_test_50 = train_test_split(
    images, labels, test_size=0.50, random_state=42, stratify=labels
)

print(f"50/50 Split - Train shape: {X_train_50.shape}, Test shape: {X_test_50.shape}")
```

```python
# Split into 80% train and 20% test
X_train_80, X_test_20, y_train_80, y_test_20 = train_test_split(
    images, labels, test_size=0.20, random_state=42, stratify=labels
)

print(f"80/20 Split - Train shape: {X_train_80.shape}, Test shape: {X_test_20.shape}")
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from sklearn.metrics import classification_report, f1_score
import matplotlib.pyplot as plt
```

```python
def create_cnn_model(input_shape):
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
        MaxPooling2D(2, 2),

        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D(2, 2),

        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.5),
        Dense(1, activation='sigmoid')  # Binary classification
    ])

    model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model
```

```python
def train_and_evaluate(X_train, y_train, X_test, y_test, title="Split"):
    model = create_cnn_model(X_train.shape[1:])

    history = model.fit(X_train, y_train,
                        epochs=10,
                        validation_data=(X_test, y_test),
                        batch_size=32,
                        verbose=1)

    # Evaluate
    loss, accuracy = model.evaluate(X_test, y_test)
    print(f"\n📊 {title} Accuracy: {accuracy:.4f}, Loss: {loss:.4f}")

    # F1 Score
    y_pred = (model.predict(X_test) > 0.5).astype("int32")
    f1 = f1_score(y_test, y_pred)
    print(f"🎯 F1 Score: {f1:.4f}")

    # Classification Report
    print("\nClassification Report:\n", classification_report(y_test, y_pred))

    # Plot
    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1)
    plt.plot(history.history['accuracy'], label='Train Acc')
    plt.plot(history.history['val_accuracy'], label='Val Acc')
    plt.title(f'{title} Accuracy')
    plt.legend()

    plt.subplot(1,2,2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.title(f'{title} Loss')
    plt.legend()
    plt.show()
```

```python
train_and_evaluate(X_train_70, y_train_70, X_test_30, y_test_30, title="70/30 Split")
```

```python
train_and_evaluate(X_train_50, y_train_50, X_test_50, y_test_50, title="📊 50/50 Split")
train_and_evaluate(X_train_80, y_train_80, X_test_20, y_test_20, title="📊 80/20 Split")
```
